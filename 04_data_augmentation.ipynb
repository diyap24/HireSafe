```python
import pandas as pd
import numpy as np
import random

# Load training data
train_df = pd.read_csv('C:/Users/diyad/Desktop/HireSafe/data/train.csv')

print(f"Original dataset: {len(train_df)} samples")
print(f"Fraudulent: {train_df['fraudulent'].sum()} ({train_df['fraudulent'].mean()*100:.1f}%)")
print(f"Legitimate: {(1-train_df['fraudulent']).sum()} ({(1-train_df['fraudulent']).mean()*100:.1f}%)")
```

```python
class SyntheticScamGenerator:
    """Generate synthetic scam job postings using templates"""
    
    def __init__(self):
        # Template components
        self.urgency_phrases = [
            "URGENT!", "ACT NOW!", "LIMITED TIME ONLY!", "IMMEDIATE START!",
            "DON'T MISS OUT!", "POSITIONS FILLING FAST!", "APPLY TODAY!"
        ]
        
        self.job_titles = [
            "Data Entry Specialist", "Customer Service Representative",
            "Virtual Assistant", "Product Tester", "Package Inspector",
            "Online Survey Taker", "Social Media Manager", "Remote Assistant"
        ]
        
        self.salary_claims = [
            "$8,000-$12,000 per week", "$5,000/week working from home",
            "$10,000+ monthly", "$500-$800 daily", "$100/hour guaranteed"
        ]
        
        self.payment_requests = [
            "Send $299 training fee via PayPal.",
            "Wire $200 for background check to get started.",
            "Pay $150 processing fee upfront.",
            "Transfer $99 for starter kit via Western Union.",
            "Refundable deposit of $250 required."
        ]
        
        self.contact_methods = [
            "Contact me on WhatsApp: +1-555-SCAM",
            "Text me at: +1-555-1234",
            "Reach out via Telegram: @scammer",
            "Email me at: jobsoffer@gmail.com",
            "Message me on my personal email"
        ]
        
        self.descriptions = [
            "Work from home! No experience needed! Flexible hours!",
            "Easy work anyone can do! Start earning today!",
            "Join our team of successful remote workers!",
            "Simple tasks, great pay! Perfect for stay-at-home parents!",
            "Be your own boss! Work whenever you want!"
        ]
    
    def generate_scam(self):
        """Generate one synthetic scam posting"""
        urgency = random.choice(self.urgency_phrases)
        title = random.choice(self.job_titles)
        salary = random.choice(self.salary_claims)
        payment = random.choice(self.payment_requests)
        contact = random.choice(self.contact_methods)
        description = random.choice(self.descriptions)
        
        full_text = f"{urgency} {title} - {salary}. {description} {payment} {contact}"
        
        return {
            'title': f"{urgency} {title}",
            'location': random.choice(['Remote', 'Work from Home', 'US, Anywhere']),
            'department': '',
            'salary_range': salary,
            'company_profile': '',  # Vague company info
            'description': full_text,
            'requirements': 'No experience required! Must be able to start immediately!',
            'benefits': 'Flexible hours! Work from anywhere! Unlimited earning potential!',
            'telecommuting': 1,
            'has_company_logo': 0,
            'has_questions': 0,
            'employment_type': 'Full-time',
            'required_experience': 'Not Applicable',
            'required_education': '',
            'industry': '',
            'function': '',
            'fraudulent': 1,  # Label as fraudulent
            'full_text': full_text
        }
    
    def generate_batch(self, n=500):
        """Generate n synthetic scams"""
        scams = []
        for _ in range(n):
            scams.append(self.generate_scam())
        return pd.DataFrame(scams)

# Test
generator = SyntheticScamGenerator()
sample_scam = generator.generate_scam()

print("Sample synthetic scam:")
print("="*80)
print(f"Title: {sample_scam['title']}")
print(f"Full text: {sample_scam['full_text']}")
print("="*80)
```

```python
# Generate 800 synthetic scams
print("Generating synthetic scams...")
synthetic_scams = generator.generate_batch(n=800)

print(f"\n✓ Generated {len(synthetic_scams)} synthetic scam postings")
print(f"\nSample titles:")
for title in synthetic_scams['title'].head(5):
    print(f"  - {title}")
```

```python
# Combine original + synthetic
augmented_train = pd.concat([train_df, synthetic_scams], ignore_index=True)

print(f"Original training data: {len(train_df)}")
print(f"Synthetic scams added: {len(synthetic_scams)}")
print(f"Augmented training data: {len(augmented_train)}")

print(f"\nNew class distribution:")
print(f"Fraudulent: {augmented_train['fraudulent'].sum()} ({augmented_train['fraudulent'].mean()*100:.1f}%)")
print(f"Legitimate: {(1-augmented_train['fraudulent']).sum()} ({(1-augmented_train['fraudulent']).mean()*100:.1f}%)")
```

```python
# Save augmented training set
augmented_train.to_csv('C:/Users/diyad/Desktop/HireSafe/data/train_augmented.csv', index=False)

print("✓ Saved train_augmented.csv")
print(f"\nYou now have:")
print(f"  - Original train: {len(train_df)} samples")
print(f"  - Augmented train: {len(augmented_train)} samples")
print(f"  - Improvement: Class balance improved from 7% to {augmented_train['fraudulent'].mean()*100:.1f}% fraud")
```
